```python
import numpy as np
import re

def touching(matrix,matrix2):
  if np.where(matrix==1)[0] == np.where(matrix2==2)[0]: # If in the same row...
    if abs(np.where(matrix==1)[1] - np.where(matrix2==2)[1]) > 1:
      return False 
    else:
      return True
  if np.where(matrix==1)[1] == np.where(matrix2==2)[1]: # If in the same column...
    if abs(np.where(matrix==1)[0] - np.where(matrix2==2)[0]) > 1:
      return False 
    else:
      return True
  if abs(np.where(matrix==1)[0] - np.where(matrix2==2)[0]) == 1: # If the difference in row is equal to one
    if abs(np.where(matrix==1)[1] - np.where(matrix2==2)[1]) == 1: # and if the difference in column is equal to one
      return True
  return False

def update_tails(matrix, matrix2):
  old_column = int(np.where(matrix2==2)[1])
  old_row = int(np.where(matrix2==2)[0])
  if np.where(matrix==1)[0]==np.where(matrix2==2)[0] and np.where(matrix==1)[1]-np.where(matrix2==2)[1]==2:
    matrix2 = moveright(matrix2,2)
  elif np.where(matrix==1)[0] == np.where(matrix2==2)[0] and np.where(matrix2==2)[1]-np.where(matrix==1)[1]==2:
    matrix2 = moveleft(matrix2,2)
  elif np.where(matrix==1)[1] == np.where(matrix2==2)[1] and np.where(matrix2==2)[0]-np.where(matrix==1)[0]==2: 
    matrix2 = moveup(matrix2,2)
  elif np.where(matrix==1)[1] == np.where(matrix2==2)[1] and np.where(matrix==1)[0]-np.where(matrix2==2)[0]==2:
    matrix2 = movedown(matrix2,2) 
  elif np.where(matrix==1)[0]-np.where(matrix2==2)[0]==1 and np.where(matrix2==2)[1]-np.where(matrix==1)[1]==2:
    matrix2[old_row+1][old_column-1] = 2
  elif np.where(matrix2==2)[0]-np.where(matrix==1)[0]==1 and np.where(matrix2==2)[1]-np.where(matrix==1)[1]==2:
    matrix2[old_row-1][old_column-1] = 2
  elif np.where(matrix==1)[0]-np.where(matrix2==2)[0]==1 and np.where(matrix==1)[1]-np.where(matrix2==2)[1]==2:
    matrix2[old_row+1][old_column+1] = 2
  elif np.where(matrix2==2)[0]-np.where(matrix==1)[0]==1 and np.where(matrix==1)[1]-np.where(matrix2==2)[1]==2:
    matrix2[old_row-1][old_column+1] = 2 
  elif np.where(matrix2==2)[0]-np.where(matrix==1)[0]==2 and np.where(matrix==1)[1]-np.where(matrix2==2)[1]==1:
    matrix2[old_row-1][old_column+1] = 2
  elif np.where(matrix2==2)[0]-np.where(matrix==1)[0]==2 and np.where(matrix2==2)[1]-np.where(matrix==1)[1]==1:
    matrix2[old_row-1][old_column-1] = 2
  elif np.where(matrix==1)[0]-np.where(matrix2==2)[0]==2 and np.where(matrix==1)[1]-np.where(matrix2==2)[1]==1:
    matrix2[old_row+1][old_column+1] = 2
  elif np.where(matrix==1)[0]-np.where(matrix2==2)[0]==2 and np.where(matrix2==2)[1]-np.where(matrix==1)[1]==1:
    matrix2[old_row+1][old_column-1] = 2
  matrix2[old_row][old_column] = 1   
  return matrix2

def increase(matrix): # Increases the matrix size 
  if np.where(matrix==1)[0]==0 or np.where(matrix==2)[0]==0: # If you need more space at the top...
    aList = [0] * matrix.shape[1]
    matrix = np.insert(matrix, 0, aList, axis=0)
  if np.where(matrix==1)[0] == matrix.shape[0]-1 or np.where(matrix==2)[0] == matrix.shape[0]-1: # If you need more spare at the bottom...
    aList = [0] * matrix.shape[1]
    matrix = np.insert(matrix, matrix.shape[0]-1, aList, axis=0)
  if np.where(matrix==1)[1] == 0 or np.where(matrix==2) == 0:
    matrix = np.c_[np.zeros([len(matrix),1]), matrix]
  if np.where(matrix==1)[1] == matrix.shape[1]-1 or np.where(matrix==2)[1] == matrix.shape[1]-1:
     matrix = np.append(matrix, np.zeros([len(matrix),1]),1)
  matrix = matrix.astype(int)  
  return matrix

def moveup(matrix,variable):
  if np.where(matrix==variable)[0] == 0:
    matrix = increase(matrix)
  old_row = int(np.where(matrix==variable)[0]) 
  matrix[old_row-1][np.where(matrix==variable)[1]] = variable 
  if variable == 1:
    matrix[old_row][np.where(matrix==variable)[1]] = 0 
  if variable == 2:
    matrix[old_row][np.where(matrix==variable)[1]] = 1    
  return matrix

def movedown(matrix,variable):
  if np.where(matrix==variable)[0] == matrix.shape[0]-1:
    matrix = increase(matrix)
  old_row = int(np.where(matrix==variable)[0]) 
  matrix[old_row+1][np.where(matrix==variable)[1]] = variable 
  if variable == 1:
    matrix[old_row][np.where(matrix==variable)[1]] = 0 
  if variable == 2:
    matrix[old_row][np.where(matrix==variable)[1]] = 1   
  return matrix

def moveright(matrix,variable):
  if np.where(matrix==variable)[1] == matrix.shape[1]-1:
    matrix = increase(matrix)  
  old_column = int(np.where(matrix==variable)[1]) 
  old_row = int(np.where(matrix==variable)[0])
  matrix[old_row][old_column+1] = variable 
  if variable == 1:
    matrix[old_row][old_column] = 0 
  if variable == 2:
    matrix[old_row][old_column] = 1
  return matrix

def moveleft(matrix, variable):
  if np.where(matrix==variable)[1] == 0:
    matrix = increase(matrix)
  old_column = int(np.where(matrix==variable)[1])
  row = int(np.where(matrix==variable)[0])
  matrix[row][old_column-1] = variable
  if variable == 1:  
    matrix[row][old_column] = 0
  if variable == 2:
    matrix[row][old_column] == 1
  return matrix

myList = [0] * 1000
matrix = [myList for _ in range(1000)]
matrix = np.array(matrix)
matrix[500][500] = 1

matrix2 = [myList for _ in range(1000)]
matrix2 = np.array(matrix2)
matrix2[500][500] = 2

with open('day_9_input.txt', mode = 'r') as myFile:
  lines = [line.rstrip() for line in myFile]
  for line in lines:
    step_size = int(re.findall('[0-9]+', line)[0])
    for i in range(0,step_size):
      if line[0] == "R":
        matrix = moveright(matrix,1)
      if line[0] == "L":
        matrix = moveleft(matrix,1)
      if line[0] == "U":
        matrix = moveup(matrix,1)
      if line[0] == "D":
        matrix = movedown(matrix,1)
      if touching(matrix,matrix2) == False:
        matrix2 = update_tails(matrix,matrix2)
print(np.sum(matrix2)-1)


    
```
