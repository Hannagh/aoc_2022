```python
import re

class TreeNode:
  def __init__(self, value, name):
    self.value = value # data
    self.children = [] # references to other nodes
    self.name = name # name of tree/branch

  def add_child(self, child_node):
    # creates parent-child relationship
    self.children.append(child_node) 
    
  def remove_child(self, child_node):
    # removes parent-child relationship
    print("Removing " + child_node.name + " from " + self.name)
    self.children = [child for child in self.children 
                     if child is not child_node]

  def traverse(self):
    nodes_to_visit = [self]
    total_of_dirs = 0
    while len(nodes_to_visit) > 0:
      current_node = nodes_to_visit.pop()
      print('Current node name ' + current_node.name + ' and value is ' + str(current_node.value))
      if current_node.value <= 100000:
        total_of_dirs += current_node.value
        print(total_of_dirs)
      nodes_to_visit += current_node.children

  def traverse2(self, list_small_dir):
    nodes_to_visit = [self]
    while len(nodes_to_visit) > 0:
      current_node = nodes_to_visit.pop()
      #print('Current node name ' + current_node.name + ' and value is ' + str(current_node.value))
      list_small_dir.append(current_node.value)
      nodes_to_visit += current_node.children      

def dfs2(stack_of_visited, tree, mylist):  # Return the value of any tree
  if tree.name not in stack_of_visited: # If you haven't visited the node yet
    stack_of_visited.add(tree.name)
  for i in tree.children:
      tree.value += dfs2(stack_of_visited, i, mylist) 
      print('Tree name is: ' + i.name)
      print('Tree value is: ' + str(i.value))
      mylist.append(i.value)
  return tree.value  

def dfs(stack_of_visited, tree, rolling_sum):  #function for dfs 
  print()
  initial_sum = rolling_sum
  if tree.name not in stack_of_visited: # If you haven't visited the node yet
    rolling_sum += tree.value 
    stack_of_visited.add(tree.name)
  outer_sum = rolling_sum
  print(initial_sum, outer_sum)
  for i in tree.children:
    inner_sum = rolling_sum
    print(initial_sum, outer_sum, inner_sum)
    dfs(stack_of_visited, i, rolling_sum) # Pass stack of visited, tree interested in

def search(stack_of_visited, tree, parent_tree):  #function for dfs 
  print('Tree name is: ' + tree.name)
  if tree.name not in stack_of_visited:
    if len(tree.children)==0:
      print('Tree name is :' + tree.name)
      print('Parent tree name is: ' + parent_tree.name)
      parent_tree.value += tree.value    
      parent_tree.remove_child(tree)
  for i in tree.children:
    search(stack_of_visited, i, tree) # Pass stack of visited, tree interested in

list_of_previous_trees = []
myTree = TreeNode(0,'/')
with open('day_7_input.txt', mode = 'r') as myfile:
  for line in myfile:
    if "$ cd /" in line:
      pointing_to = myTree
    if re.findall("\d+", line): 
      pointing_to.value += int(re.findall("\d+", line)[0])
    if 'dir ' in line:
      newDir = TreeNode(0, re.findall('(?<=dir ).*', line)[0]) # Create a new node
      pointing_to.add_child(newDir) # Add it to the new tree
    if '$ cd ' in line and '$ cd ..' not in line: # If changing a directory going downward
      list_of_previous_trees.append(pointing_to) # Add to the list of previous trees    
      name_to_change_to = re.findall('(?<=cd ).*', line)[0] 
      for i in pointing_to.children:
        if i.name == name_to_change_to: # Find node that corresponds to name_to_change_to
          pointing_to = i 
    if '$ cd ..' in line: # If moving up a level, search myTree where name == parent
      pointing_to = list_of_previous_trees.pop()

stack_of_visited = set()
myList = []
dfs2(stack_of_visited, myTree, myList)

#myTree.traverse()
#print('Move to part 2')

anotherlist = []
#myTree.traverse2(mylist)

for i in myList:
  if i >= 8381165:
    anotherlist.append(i)
print(myList)
print(min(anotherlist))

#stack_of_visited = set()
#search(stack_of_visited, myTree, parentTree)

# Search through myTree to find directories with values below 100,000
#for i in myTree.children:
#  print(i.value)   
#  print(i.children)   
```
