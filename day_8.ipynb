```python
import copy

#See every tree, as long as there isn't a tree larger than their own tree in front of it

def HighestTree(aList,aValue):
  numTrees = 1
  for index,value in enumerate(aList):
    if index==0:
      continue
    if aList[index-1] < aValue:
      numTrees += 1
      continue
    if aList[index-1] >= aValue:
      return numTrees
  return numTrees

myList = []
anotherList = []
with open('day_8_input.txt', mode='r') as myFile: 
  for line in myFile:
    for i in line:
      if i != '\n':
        anotherList.append(i)
      else:
        myList.append(anotherList)
        anotherList = []

seenTrees = []
for i in range(len(myList)):
  zeros = [0] * len(myList[0])
  seenTrees.append(zeros)

#Looking east
for index2, values2 in enumerate(myList):
  seenTrees[index2][0] = 1 # See trees that are along the edge
  highest_tree = myList[index2][0]
  highest_tree_location = 0
  for index,value in enumerate(myList[index2]):
    if value > highest_tree:
      highest_tree_location = index
      highest_tree = value
      seenTrees[index2][highest_tree_location] = 1
  seenTrees[index2][highest_tree_location] = 1

# Looking west
for sub in myList:
     sub.reverse()
for sub in seenTrees:
     sub.reverse()     
for index2, values2 in enumerate(myList):
  seenTrees[index2][0] = 1 # See trees that are along the edge
  highest_tree = myList[index2][0]
  highest_tree_location = 0
  for index,value in enumerate(myList[index2]):
    if value > highest_tree:
      highest_tree_location = index
      highest_tree = value
      seenTrees[index2][highest_tree_location] = 1
  seenTrees[index2][highest_tree_location] = 1

#Looking north
myList = list(map(list, zip(*myList)))
for sub in myList:
     sub.reverse()
seenTrees = list(map(list, zip(*seenTrees)))
for sub in seenTrees:
     sub.reverse()     
for index2, values2 in enumerate(myList):
  seenTrees[index2][0] = 1 # See trees that are along the edge
  highest_tree = myList[index2][0]
  highest_tree_location = 0
  for index,value in enumerate(myList[index2]):
    if value > highest_tree:
      highest_tree_location = index
      highest_tree = value
      seenTrees[index2][highest_tree_location] = 1
  seenTrees[index2][highest_tree_location] = 1

#Looking south
for sub in myList:
  sub.reverse()  
for sub in seenTrees:
  sub.reverse()        
for index2, values2 in enumerate(myList):
  seenTrees[index2][0] = 1 # See trees that are along the edge
  highest_tree = myList[index2][0]
  highest_tree_location = 0
  for index,value in enumerate(myList[index2]):
    if value > highest_tree:
      highest_tree_location = index
      highest_tree = value
      seenTrees[index2][highest_tree_location] = 1
  seenTrees[index2][highest_tree_location] = 1

# Part 2
myList = []
anotherList = []
with open('day_8_input.txt', mode='r') as myFile: 
  for line in myFile:
    for i in line:
      if i != '\n':
        anotherList.append(i)
      else:
        myList.append(anotherList)
        anotherList = []

scenicScores = []
for i in range(len(myList)):
  zeros = [0] * len(myList[0])
  scenicScores.append(zeros)

# if zero is in the index, don't search a certian direction
for i, value in enumerate(myList):
  for j,value2 in enumerate(value):
    if i == 0 or j == 0 or i == len(myList)-1 or j==len(myList[i])-1:
      scenicScore = 0
    else:    
      changingList = copy.deepcopy(myList)
      east_sub_list = changingList[i][j+1:]
      EastScore = HighestTree(east_sub_list,value2)
      for sub in changingList:
          sub.reverse()
      west_sub_list = changingList[i][len(changingList)-j:]
      WestScore = HighestTree(west_sub_list,value2) 
      changingList = list(map(list, zip(*changingList)))
      for sub in changingList:
          sub.reverse()
      changingList.reverse()
      north_sub_list = changingList[j][len(changingList)-i:]
      NorthScore = HighestTree(north_sub_list,value2)
      for sub in changingList:
          sub.reverse()
      south_sub_list = changingList[j][i+1:]
      SouthScore = HighestTree(south_sub_list,value2)
      scenicScore = NorthScore * SouthScore * EastScore * WestScore
    scenicScores[i][j] = scenicScore
max_values = [max(i) for i in scenicScores]
print(max(max_values))


```
