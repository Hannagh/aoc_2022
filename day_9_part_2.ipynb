```python
import numpy as np
import re

def touching(matrix,matrix2, heads_var, tails_var):
  heads_column = int(np.where(matrix==heads_var)[1])
  heads_row = int(np.where(matrix==heads_var)[0])
  tails_column = int(np.where(matrix2==tails_var)[1])
  tails_row = int(np.where(matrix2==tails_var)[0])
  if heads_row == tails_row: # If in the same row...
    if abs(heads_column - tails_column) > 1:
      return False 
    else:
      return True
  if heads_column == tails_column: # If in the same column...
    if abs(heads_row - tails_row) > 1:
      return False 
    else:
      return True
  if abs(heads_row - tails_row) == 1: # If the difference in row is equal to one
    if abs(heads_column - tails_column) == 1: # and if the difference in column is equal to one
      return True
  return False

def update_tails(matrix, matrix2, heads_var, tails_var):
  old_column = int(np.where(matrix2==tails_var)[1])
  old_row = int(np.where(matrix2==tails_var)[0])
  heads_column = int(np.where(matrix==heads_var)[1])
  heads_row = int(np.where(matrix==heads_var)[0])
  if heads_row==np.where(matrix2==tails_var)[0] and heads_column-old_column==2: # If in the same row 
    matrix2 = moveright(matrix2,tails_var)
  elif heads_row == old_row and old_column-heads_column==2: # If in the same row
    matrix2 = moveleft(matrix2,tails_var)
  elif heads_column == old_column and old_row-heads_row==2: # If in the same column
    matrix2 = moveup(matrix2,tails_var)
  elif heads_column == old_column and heads_row-old_row==2: # If in the same column
    matrix2 = movedown(matrix2,tails_var) 
  elif heads_row-old_row==1 and old_column-heads_column==2:
    matrix2[old_row+1][old_column-1] = tails_var
  elif old_row-heads_row==1 and old_column-heads_column==2:
    matrix2[old_row-1][old_column-1] = tails_var
  elif heads_row-old_row==1 and heads_column-old_column==2:
    matrix2[old_row+1][old_column+1] = tails_var
  elif old_row-heads_row==1 and heads_column-old_column==2:
    matrix2[old_row-1][old_column+1] = tails_var 
  elif old_row-heads_row==2 and heads_column-old_column==1:
    matrix2[old_row-1][old_column+1] = tails_var
  elif old_row-heads_row==2 and old_column-heads_column==1:
    matrix2[old_row-1][old_column-1] = tails_var
  elif heads_row-old_row==2 and heads_column-old_column==1:
    matrix2[old_row+1][old_column+1] = tails_var
  elif heads_row-old_row==2 and old_column-heads_column==1:
    matrix2[old_row+1][old_column-1] = tails_var
  elif heads_row-old_row==2 and heads_column-old_column==2: # head is in a higher row, and head is in a righter row
    matrix2[old_row+1][old_column+1] = tails_var
  elif heads_row-old_row==2 and old_column-heads_column==2: # head is in a higher row, and head is in a lefter row
    matrix2[old_row+1][old_column-1] = tails_var
  elif old_row-heads_row==2 and heads_column-old_column==2: # head is in a lower row, and head is in a righter row
    matrix2[old_row-1][old_column+1] = tails_var
  elif old_row-heads_row==2 and old_column-heads_column==2: # head is in a lower row, and head is in a lefter row
    matrix2[old_row-1][old_column-1] = tails_var        
  if tails_var == 10:
    matrix2[old_row][old_column] = 1
  else:
    matrix2[old_row][old_column] = 0   
  return matrix2

def increase(matrix, variable): # Increases the matrix size 
  if np.where(matrix==variable)[0]==0: # If you need more space at the top...
    aList = [0] * matrix.shape[1]
    matrix = np.insert(matrix, 0, aList, axis=0)
  if np.where(matrix==variable)[0] == matrix.shape[0]-1: # If you need more spare at the bottom...
    aList = [0] * matrix.shape[1]
    matrix = np.insert(matrix, matrix.shape[0]-1, aList, axis=0)
  if np.where(matrix==variable)[1] == 0 or np.where(matrix==variable) == 0:
    matrix = np.c_[np.zeros([len(matrix),1]), matrix]
  if np.where(matrix==variable)[1] == matrix.shape[1]-1 or np.where(matrix==variable)[1] == matrix.shape[1]-1:
     matrix = np.append(matrix, np.zeros([len(matrix),1]),1)
  matrix = matrix.astype(int)  
  return matrix

def moveup(matrix,variable):
  if np.where(matrix==variable)[0] == 0:
    matrix = increase(matrix, variable)
  old_row = int(np.where(matrix==variable)[0]) 
  matrix[old_row-1][np.where(matrix==variable)[1]] = variable 
  if variable == 1:
    matrix[old_row][np.where(matrix==variable)[1]] = 0 
  if variable == 2:
    matrix[old_row][np.where(matrix==variable)[1]] = 1    
  return matrix

def movedown(matrix,variable):
  if np.where(matrix==variable)[0] == matrix.shape[0]-1:
    matrix = increase(matrix, variable)
  old_row = int(np.where(matrix==variable)[0]) 
  matrix[old_row+1][np.where(matrix==variable)[1]] = variable 
  if variable == 1:
    matrix[old_row][np.where(matrix==variable)[1]] = 0 
  if variable == 2:
    matrix[old_row][np.where(matrix==variable)[1]] = 1   
  return matrix

def moveright(matrix,variable):
  if np.where(matrix==variable)[1] == matrix.shape[1]-1:
    matrix = increase(matrix, variable)  
  old_column = int(np.where(matrix==variable)[1]) 
  old_row = int(np.where(matrix==variable)[0])
  matrix[old_row][old_column+1] = variable 
  if variable == 1:
    matrix[old_row][old_column] = 0 
  if variable == 2:
    matrix[old_row][old_column] = 1
  return matrix

def moveleft(matrix, variable):
  if np.where(matrix==variable)[1] == 0:
    matrix = increase(matrix, variable)
  old_column = int(np.where(matrix==variable)[1])
  row = int(np.where(matrix==variable)[0])
  matrix[row][old_column-1] = variable
  if variable == 1:  
    matrix[row][old_column] = 0
  if variable == 2:
    matrix[row][old_column] == 1
  return matrix

myList = [0] * 1000
matrix = [myList for _ in range(1000)]
matrix = np.array(matrix)
matrix[500][500] = 1

matrix2 = [myList for _ in range(1000)]
matrix2 = np.array(matrix2)
matrix2[500][500] = 2

matrix3 = [myList for _ in range(1000)]
matrix3 = np.array(matrix3)
matrix3[500][500] = 3

matrix4 = [myList for _ in range(1000)]
matrix4 = np.array(matrix4)
matrix4[500][500] = 4

matrix5 = [myList for _ in range(1000)]
matrix5 = np.array(matrix5)
matrix5[500][500] = 5

matrix6 = [myList for _ in range(1000)]
matrix6 = np.array(matrix6)
matrix6[500][500] = 6

matrix7 = [myList for _ in range(1000)]
matrix7 = np.array(matrix7)
matrix7[500][500] = 7

matrix8 = [myList for _ in range(1000)]
matrix8 = np.array(matrix8)
matrix8[500][500] = 8

matrix9 = [myList for _ in range(1000)]
matrix9 = np.array(matrix9)
matrix9[500][500] = 9

matrix10 = [myList for _ in range(1000)]
matrix10 = np.array(matrix10)
matrix10[500][500] = 10

with open('day_9_input.txt', mode = 'r') as myFile:
  lines = [line.rstrip() for line in myFile]
  for line in lines:
    step_size = int(re.findall('[0-9]+', line)[0])
    for i in range(0,step_size):
      if line[0] == "R":
        matrix = moveright(matrix,1)
      if line[0] == "L":
        matrix = moveleft(matrix,1)
      if line[0] == "U":
        matrix = moveup(matrix,1)
      if line[0] == "D":
        matrix = movedown(matrix,1)
      if touching(matrix, matrix2, 1, 2) == False:
        matrix2 = update_tails(matrix,matrix2, 1, 2)
      if touching(matrix2, matrix3, 2, 3) == False:   
        matrix3 = update_tails(matrix2,matrix3, 2, 3)
      if touching(matrix3,matrix4, 3, 4) == False:  
        matrix4 = update_tails(matrix3,matrix4, 3, 4)
      if touching(matrix4, matrix5, 4, 5) == False:
        matrix5 = update_tails(matrix4,matrix5, 4, 5)
      if touching(matrix5,matrix6, 5, 6) == False:  
        matrix6 = update_tails(matrix5,matrix6, 5, 6)
      if touching(matrix6,matrix7, 6, 7) == False:        
        matrix7 = update_tails(matrix6,matrix7, 6, 7)
      if touching(matrix7, matrix8, 7, 8) == False:   
        matrix8 = update_tails(matrix7,matrix8, 7, 8)
      if touching(matrix8,matrix9, 8, 9) == False:        
        matrix9 = update_tails(matrix8,matrix9, 8, 9)
      if touching(matrix9,matrix10, 9, 10) == False:
        matrix10 = update_tails(matrix9, matrix10, 9, 10)      
print(np.sum(matrix10)-9)
```
